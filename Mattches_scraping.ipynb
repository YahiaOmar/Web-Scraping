```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


Date = '10/6/2024'
url = f"https://www.yallakora.com/match-center/%D9%85%D8%B1%D9%83%D8%B2-%D8%A7%D9%84%D9%85%D8%A8%D8%A7%D8%B1%D9%8A%D8%A7%D8%AA?date={Date}#"
response = requests.get(url)
```

```python

# Step 2: Check if the request was successful
if response.status_code == 200:
    print("Successfully connected to BBC News.")
else:
    print(f"Failed to connect, status code: {response.status_code}")

```

```python
# Step 3: Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Step 4: Find the elements that contain the article titles
# BBC news articles are usually in <h3> tags with class 'gs-c-promo-heading__title'
championships = soup.find_all("div", class_="matchCard")
```

```python
list_of_cmp_titles = []
teamsA = []
teamsB = []
scores = []
info = []
times = []
for i in range(len(championships)):
    cmp_title = championships[i].contents[1].find("h2").get_text().strip()
    list_of_cmp_titles.append(cmp_title)

    all_matches = championships[i].contents[3].find_all("div", class_='liItem')

    for n in range(len(all_matches)):
        cmp_title = championships[i].contents[1].find("h2").get_text().strip()
        list_of_cmp_titles.append(cmp_title)


        teamA = all_matches[n].find("div", class_='teamA').get_text().strip()
        teamB = all_matches[n].find("div", class_='teamB').get_text().strip()

        match_result = all_matches[n].find("div", class_='MResult').find_all('span', class_='score')
        score = f"{match_result[0].get_text().strip()} - {match_result[1].get_text().strip()}"

        time = all_matches[n].find("div", class_='MResult').find("span", class_='time').get_text().strip()

        info.append([cmp_title,teamA,teamB,score,time])




print(times)
print(teamsA)
print(info)
df = pd.DataFrame(info, columns=['chmp','teamB','teamA', 'score','time'])
```
